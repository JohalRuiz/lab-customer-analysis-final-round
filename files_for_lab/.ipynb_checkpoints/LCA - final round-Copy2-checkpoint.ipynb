### BONUS: Put all the previously mentioned data transformations into a function.

```python
df_function.columns
```

```python
import time
from datetime import date
```

```python
def standard_clean(df_function):
    df_function.columns = [columns.lower().replace(' ', '_') for column in df_function.columns]
    df_function_num = df_function.select_dtypes(include='number')
    df_function_cat = df_function.select_dtypes(include='object')
    if df_function.isnull().values.any():
        df_function = df_function.dropna(how='any',axis=0) 
    else:
        df_function = df_function
    df_function['effective_to_month'] = df_function['effective_to_date'].dt.to_period('M')
    df_function = df_function.sort_values(by=['effective_to_date'], ascending = True)
    df_function = df_function.reset_index(drop=True)
    return standard_clean
```

### Now, let's check data correlation WITH normalization BEFORE X-y split:

```python
df4 = df3.copy()
```

```python
#normalization
```

```python
def boxcox_transform(df4):
    numeric_cols = df4.select_dtypes(np.number).columns
    _ci = {column: None for column in numeric_cols}
    for column in numeric_cols:
        # since i know any columns should take negative numbers, to avoid -inf in df
        df4[column] = np.where(df4[column]<=0, np.NAN, df4[column]) 
        df4[column] = df4[column].fillna(df4[column].mean())
        transformed_data, ci = stats.boxcox(df4[column])
        df4[column] = transformed_data
        _ci[column] = ci #ci means confidence interval
    return df4, _ci
```

```python
df4, _ci = boxcox_transform(df4) #(here the function is applied to the dataframe and then _ci displayed)
_ci
```

```python
df4.shape
```

```python
#sns.pairplot(df4.select_dtypes(np.number))
#plt.show()
```

```python
def remove_outliers(df4, threshold=1.5, in_columns=df4.select_dtypes(np.number).columns, skip_columns=[]):
    for column in in_columns:
        if column not in skip_columns:
            upper = np.percentile(df4[column],75)
            lower = np.percentile(df4[column],25)
            iqr = upper - lower
            upper_limit = upper + threshold * iqr
            lower_limit = lower - threshold * iqr
            df4 = df4[(df[column]>lower_limit) & (df4[column]<upper_limit)]
    return df4
```

```python
df4 = remove_outliers(df4, threshold=1.5, in_columns=['customer_lifetime_value', 'income', 'monthly_premium_auto', 'total_claim_amount'])
# This function is applied ONLY to continuous values
df4
```
